```python
from geopy.geocoders import Nominatim
import requests

def get_coordinates(address):
    """
    Convertit une adresse en latitude et longitude en utilisant Nominatim.
    """
    geolocator = Nominatim(user_agent="osm_geocoder")
    location = geolocator.geocode(address)
    if location:
        return location.latitude, location.longitude
    else:
        raise Exception("Adresse introuvable.")

def get_nearby_places(lat, lon, radius=1000):
    """
    Récupère les lieux dans un rayon donné autour de coordonnées géographiques
    en utilisant Overpass API, tout en filtrant les rues.
    """
    overpass_url = "http://overpass-api.de/api/interpreter"
    query = f"""
    [out:json];
    (
      node["amenity"](around:{radius},{lat},{lon});
      way["amenity"](around:{radius},{lat},{lon});
      node["tourism"](around:{radius},{lat},{lon});
      way["tourism"](around:{radius},{lat},{lon});
      node["leisure"](around:{radius},{lat},{lon});
      way["leisure"](around:{radius},{lat},{lon});
    );
    out center;
    """
    response = requests.get(overpass_url, params={'data': query})

    if response.status_code == 200:
        data = response.json()
        places = []
        for element in data['elements']:
            if 'tags' in element and 'name' in element['tags']:
                name = element['tags']['name']
                places.append(name)
        return places
    else:
        raise Exception(f"Erreur avec Overpass API : {response.status_code}")

# Programme principal
if __name__ == "__main__":
    # Adresse d'entrée
    address = "58 rue du Louvre, Paris, France"
    radius = 20  # Rayon en mètres

    try:
        # Étape 1 : Obtenir les coordonnées géographiques de l'adresse
        latitude, longitude = get_coordinates(address)
        print(f"Coordonnées de '{address}': Latitude={latitude}, Longitude={longitude}")

        # Étape 2 : Rechercher les lieux dans le rayon défini
        places = get_nearby_places(latitude, longitude, radius)
        print("\nLieux trouvés :")
        for place in places:
            print(f"- {place}")
    except Exception as e:
        print(e)
```
